```python
cross_tab1 = pd.crosstab(df['channel_sales'], df['has_gas'])
cross_tab1
```

```python
cross_tab2 = pd.crosstab(df['channel_sales'], df['origin_up'])
cross_tab2
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['channel_sales'], df['has_gas'])
chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Chi-square: {chi2}\nP-value: {p}")
#statistical sign(p<0.05)
```
